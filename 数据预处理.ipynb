# 数据预处理

```python
## 对附件三数据预处理
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = "SimHei"  ## 设置中文显示
plt.rcParams['axes.unicode_minus']=False    # 默认是使用Unicode负号，设置正常显示字符，如正常显示负号
```

```python
# 读取数据
df = pd.read_excel('D:\桌面\数据挖掘-汽油辛烷值损失模型优化\暑期模拟-3-2022-08-18\模拟练习-3 B题 汽油辛烷值建模\附件三：285号和313号样本原始数据.xlsx',sheet_name="操作变量")
df.info()
```

```python
# 将数据中的空值0替换成NaN
df.replace(0, np.nan,inplace=True)
```

```python
#重复数据清洗
df.drop_duplicates(keep="first",inplace=True)
df.info()
```

```python
# 分别选择285号和313号数据
df1=df.iloc[0:39,:] # 285
df2=df.iloc[40:79,:] #313
df1.info()
```

```python
#缺失值清洗（NaN,删除全为NaN的整列数据              axis=0；how="any"--默认值)
df1 = df1.copy()  # 会创建一个新的对象进行拷贝,修改这份拷贝不会对原有对象产生影响
df2 = df2.copy()
df1.dropna(axis=1,how="all",inplace=True)
df2.dropna(axis=1,how="all",inplace=True)
```

```python
# 更新索引（drop=True: 把原来的索引index列删除，重置index)
df1.reset_index(drop=True,inplace=True)
df2.reset_index(drop=True,inplace=True)
```

```python
# 删除缺测值过多的列
df2.drop(columns=["3.0步骤FIC2432.SP","D-106热氮气流量"],inplace=True,axis=1)
```

```python
# 查找所有带NaN值的数据
x = df2[df2.isnull().values==True] # 查找所有带NaN值的数据
x
```

```python
# 查找NaN所在的行和列的位置
position1 =df2.isnull().stack()[lambda x: x].index.tolist()  
print(position1)
```

```python
# 对于存在缺测值但不多的列，用一列的平均值来填充缺测值
df2["D-121含硫污水排量"].fillna(df2["D-121含硫污水排量"].mean(),inplace=True)
df2["Unnamed: 175"].fillna(df2["Unnamed: 175"].mean(),inplace=True)
df2["Unnamed: 109"].fillna(df2["Unnamed: 109"].mean(),inplace=True)
```

```python
# 判断此时df1, df2 中是否还有NaN
### 如果你将它df.isnull().any() ，你只能找到具有NaN值的列
### 还有一个.any()会告诉你上面的任何一个是否为True
df1.isnull().any().any()
df2.isnull().any().any()
```

```python
# 获取DataFrame中NaN值的总数
df2.isnull().sum().sum()


###### 现在已经处理完样本中数据为空值的位点
```

```python
##### 最大最小的限幅方法剔除一部分不在此范围的样本
# 拉依达准则(3σ)--- 标准差σ
# 剔除超过3σ的偏离点
# vi: 剩余误差   x: 算术平均值   
# |vi|=|xi|-|x|>3σ
```

```python
# 计算每列的算数平均值
# 计算前删除含有其他非数值的列，否则会有警告
df1=df1.drop(labels="中文名称",axis=1,inplace=False)
df2=df2.drop(labels="中文名称",axis=1,inplace=False)
mean1=df1.mean(axis=0)  # 默认 axis=0 计算每列的算数平均数
mean2=df2.mean(axis=0)
```

```python
# 计算每一列的标准差
std1=df1.std(axis=0)
std2=df2.std(axis=0)

# 计算每一列的剩余误差
residualError1 = abs(df1 - mean1)
residualError2 = abs(df2 - mean2)
```

```python
# 筛选大于3σ的异常值
# 如果判断为真,则保持原来的值,否则替换为other（other默认为NaN)
df1.where(residualError1<=3*std1,inplace=True)
df2.where(residualError2<=3*std2,inplace=True)
```

```python
# 查看313号数据中异常值个数
df2.isnull().sum().sum()
```

```python
# 查找NaN所在的行和列的位置
position_df1 =residualError1.isnull().stack()[lambda x: x].index.tolist()  
print(position_df1)
position_df2 =residualError2.isnull().stack()[lambda x: x].index.tolist()  
print(position_df2)
```

```python
# 均值填补法来将处理后的285号和313号数据填入表格
mean1=df1.mean(axis=0)  
mean2=df2.mean(axis=0)
type(mean1)
```

```python
# 将数据存入csv
mean1.to_csv('D:\桌面\q285_mean.csv',index=False)
mean2.to_csv('D:\桌面\q313_mean.csv',index=False)
```
